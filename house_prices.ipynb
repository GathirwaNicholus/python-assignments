### Variance inflation factor
- high vif = multicorlinearity 
- vif can be thought of as a measure of multicorlinearity(relationship between independent variables)
- multicorlinearity of less than 5 is accepted but more than 5 is not and more than 10 is worst

```python
y = house_prices_data['price']
x = house_prices_data.drop(['price'],axis = 1)
x.head()
```

```python
vif = variance_inflation_factor
```

```python
x = add_constant(x)

#computing vif
vif_data = pd.DataFrame()
vif_data['FEATURE'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif_data.head()
```

```python
print (vif_data)
```

#### hence there is no multicorlinearity as all are 1


### Hypothesis testing using anova

```python
house_prices_data.columns
```

```python
formula = 'price ~ bedrooms + bathrooms + sqft_living + sqft_lot + floors + year_built + year_renovated + lat + long + C(waterfront) + C(view) + C(grade)'
lm = ols(formula, house_prices_data).fit()
table = sm.stats.anova_lm(lm, typ=2)
table
```

```python
# feature engineering.
# how the year renovated column looks like
house_prices_data["year_renovated"].value_counts()
```

```python
# adding a column, year since rennovation-> current date - renovated date
#1. convert the date to date_time.
#house_prices_data["year_renovated"] = pd.to_datetime(house_prices_data["year_renovated"])
house_prices_data.info()
```

```python
#current year
current_year = pd.Timestamp.now().year

#current year - year rennovated
house_prices_data["year_since_rennovation"] = current_year - house_prices_data["year_renovated"]

house_prices_data["year_since_rennovation"] = house_prices_data["year_since_rennovation"].where(house_prices_data["year_renovated"]>0, 0)
house_prices_data.head()

#, 0: This is the second argument to the where() method. It specifies that for any row where the condition (house_prices_data["year_renovated"] > 0) is False, the value in the "year_since_rennovation" column should be replaced with 0. 
```

```python
correlations1 = house_prices_data.corr()

#correlation heat map
plt.figure(figsize= (12,12))
sns.heatmap(data=correlations1, cmap= "coolwarm", annot= True)
```

```python
#checking for colinearity
x = house_prices_data.drop('price', axis=1)
x = add_constant(x)

#computing vif
vif_data = pd.DataFrame()
vif_data['FEATURE'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]

print(vif_data)
```

```python
# notice that year rennovated and year since rennovation appear to be coliner 3.9 and 3.9 respectively(however this is just for studying purposes) - recall mulitcolinearity of less than 5 is acceptable.
# Thus we have to drop 1 column
x = x.drop("year_renovated", axis=1)

#computing vif
vif_data = pd.DataFrame()
vif_data['FEATURE'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]

print(vif_data)

#now notice that year_since_renovation is down to 1.
```

```python
# data normalization
from sklearn.preprocessing import MinMaxScaler


min_max_scaler = MinMaxScaler()
data_scaled = min_max_scaler.fit_transform(x)
scaled_df = pd.DataFrame(data_scaled)
scaled_melted = scaled_df.melt(var_name= "Variable", value_name= "Scaled Value")

plt.figure(figsize=(10, 6))
sns.boxplot(x="Variable", y= "Scaled Value", data=scaled_melted, palette="viridis")
plt.title("Boxplots on Normalized sclae (0-1)")
plt.show()

```

```python
price_normalized = np.log(y)
y_transformed = pd.Series(price_normalized)
print(y_transformed)
```

```python
# ols model
model = sm.OLS(price_normalized, scaled_df).fit()
print(model.summary())
```

```python
# Plot partial regression plots to visualize linearity with each variable
fig = plt.figure(figsize=(12, 10))
plot_partregress_grid(model, fig=fig)
plt.tight_layout()
plt.show()
```

```python
# Perform Rainbow test for linearity
rainbow_statistic, rainbow_p_value = linear_rainbow(model)
rainbow_statistic, rainbow_p_value
```

The Rainbow test shows no evidence of nonlinearity (p = 0.459 > 0.05).
The partial regression plots above confirm this visually  most relationships between predictors and price are roughly linear, though a few may show mild curvature or outliers worth further exploration.

The model is linear in nature, and the linear regression assumptions hold reasonably well..
